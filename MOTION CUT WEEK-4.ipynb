```python
import time

def introduction():
    print("Welcome to the Text Adventure Game!")
    time.sleep(1)
    print("You find yourself in a dark, mysterious place.")
    time.sleep(1)
    print("Your goal is to navigate through the challenges and reach the treasure.")
    time.sleep(1)

def choose_path():
    print("\nChoose your path:")
    print("1. Go left")
    print("2. Go right")
    choice = input("Enter your choice (1 or 2): ")
    return choice

def left_path():
    print("\nYou chose the left path.")
    time.sleep(1)
    print("You encounter a giant spider!")
    time.sleep(1)
    print("What do you want to do?")
    print("1. Fight the spider")
    print("2. Try to run away")
    choice = input("Enter your choice (1 or 2): ")
    if choice == '1':
        print("\nYou defeated the spider and continue your journey.")
        return True
    else:
        print("\nThe spider was too fast, and you got caught.")
        return False

def right_path():
    print("\nYou chose the right path.")
    time.sleep(1)
    print("You find a bridge guarded by a troll.")
    time.sleep(1)
    print("What do you want to do?")
    print("1. Try to sneak past the troll")
    print("2. Challenge the troll to a riddle contest")
    choice = input("Enter your choice (1 or 2): ")
    if choice == '1':
        print("\nYou successfully sneak past the troll and continue your journey.")
        return True
    else:
        print("\nThe troll is impressed with your riddles and lets you pass.")
        return True

def main():
    introduction()

    # Start the game loop
    while True:
        choice = choose_path()

        if choice == '1':
            result = left_path()
        elif choice == '2':
            result = right_path()
        else:
            print("Invalid choice. Please enter 1 or 2.")
            continue

        if result:
            print("Congratulations! You reached the treasure and won the game.")
            break
        else:
            print("Game over. Try again.")
            break

if __name__ == "__main__":
    main()
```
